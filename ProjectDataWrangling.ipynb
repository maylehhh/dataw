# Use to import necessary libraries/packages

```python
import pandas as pd
import country_converter as cc
coco = cc.CountryConverter()
```

# Use to import files

```python
file1 = 'refugee population by country of asylum.csv'
file2 = 'gdp_csv.csv'
file3 = 'crim_off_cat.tsv'
file4 = 'secondary enrollment.csv'
file5 = 'tertiary enrollment.csv'
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3, sep='\t|\,', engine='python')
df3 = df3.rename(columns={'geo\\time':'Country Code'})
df4 = pd.read_csv(file4)
df5 = pd.read_csv(file5)
```

# Use to filter country codes

```python
url = 'https://www.countrycallingcodes.com/iso-country-codes/europe-codes.php'
countries = pd.read_html(url)
countries = countries[1]
countries = countries.drop([0,1,2])
countries_2 = countries[1]
countries_2 = countries_2.to_frame()
countries_3 = countries[2]
countries_3 = countries_3.to_frame()
```

# Use to merge files

```python
refugees_and_gdp = df2.merge(df1, left_on=['Country Name','Year', 'Country Code'], right_on=['Country Name', 'Year', 'Country Code'])
refugees_and_gdp = refugees_and_gdp.rename(columns={'Value_x':'GDP', 'Value_y':'Refugee population'})
refugees_gdp_secondary = refugees_and_gdp.merge(df4, left_on=['Country Name','Year', 'Country Code'], right_on=['Country Name', 'Year', 'Country Code'])
refugees_gdp_secondary = refugees_gdp_secondary.rename(columns={'Value':'Secondary education %gross'})
refugees = refugees_gdp_secondary.merge(df5, left_on=['Country Name','Year', 'Country Code'], right_on=['Country Name', 'Year', 'Country Code'])
refugees = refugees.rename(columns={'Value':'Tertiary education %gross'})
```

# Filter countries:

```python
refugees = refugees.merge(countries_3, left_on='Country Code', right_on=2)
df3 = df3.merge(countries_2, left_on = 'Country Code', right_on=1)
df3 = df3.drop(1, axis=1)
```

# Convert Country Codes of df3 to ISO3

```python
df3['Country Code'] = coco.convert(names = list(df3['Country Code']), src='ISO2', to='ISO3')
```
